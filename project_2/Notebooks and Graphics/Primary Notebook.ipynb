```python
import numpy as np
import pandas as pd
import random 
import matplotlib.pyplot as plt
import scipy.interpolate as sci
import seaborn as sns

global_temperatures = pd.read_csv('GLB.Ts+dSST.csv')
global_temperatures.columns = ['Year', "Jan", "Feb","Mar","Apr","May","June","July","August","September","October","November","December",'14','15','16','17', '18', '19']
g1 = global_temperatures.drop([0])
for i in range(14,20):
     g1 = g1.drop(str(i), axis = 1)
g2 = g1.copy()
g2 = g2.drop("Year", axis = 1)
g2 = g2.replace("***", np.nan)
g2 = g2.astype(float)
g2["Mean for the Year"] = g2.mean(axis = 1)
g1.join(g2["Mean for the Year"])

nl = '\n'
x = list(g1['Year'])
y = list(g2['Mean for the Year'])
g3 = pd.DataFrame({"x": x, "y": y})
xmin = g1['Year'].min()
xmax = g1['Year'].max()
plt.figure(figsize=(20,10))
plt.xticks(np.arange(0, 150, step=10))
plt.rc('font', size = 15)
plt.xlabel("Year", fontsize = 36, fontname = "Times New Roman")
plt.ylabel(f"Average Change per Year {nl} Global Temperature (Celsius)", fontsize = 36, fontname = "Times New Roman")
plt.title(f"Annual Change of Earth's Global Temperature {nl} 1880-2020", fontsize = 36, fontname = "Times New Roman")
plt.grid()
plt.plot(x,y)
plt.savefig('global_warming.jpeg')
plt.show()


```

```python
nl = "\n"
###### Creates Graphs for USA States
class USAStateGraph:
    def __init__(self, name = None, code = None, df = None):
        self.name = name
        self.code = code
        self.df = df
    def create_graph(self, name = None, code = None):
        State_Name = str(self.name)
        State_Code = str(self.code)
        self.df = pd.read_excel('PublicData_USA_States_v6 (1) - Updated.xlsx', sheet_name = State_Code, usecols=("D,F,M,T,S, L, H, G"))
        self.df.columns = ["Time Since January 1st", "A", "B", "C", "D", "E", "F", "G"]
        pd.set_option('display.max_rows', self.df.shape[0] + 1)
        pd.set_option('display.max_columns', self.df.shape[0] + 1)
        plt.title(f"Change in CO2 emissions per day in {State_Name} since January 1st, 2020")
        plt.xlabel("Time Since January 01, 2020 (Days)")
        plt.ylabel("Change in CO2 Emissions (Metric Tons)")
        columns = ["A", "B", "C", "D", "E", "F", "G"]
        for i in columns:
            plt.plot("Time Since January 1st", f'{i}', data = self.df, color = 'black')
        #plt.plot('Time Since January 1st', 'A', data=self.df, marker='o', markerfacecolor='red', markersize=2, color='red', linewidth=2, linestyle = 'dashed')
        #plt.plot('Time Since January 1st', 'Total CO2, Median', data=self.df, marker='o', markerfacecolor = 'orange', markersize = 2, color='orange', linewidth=2)
        #plt.plot('Time Since January 1st', 'Total CO2, High', data=self.df, marker='o', markerfacecolor = 'green', markersize = 2, color='green', linewidth=2, linestyle = 'dashed')
        plt.grid()
        plt.show()
    def __str__ (self, name = None):
        print(self.name)
        print(self.code)


###### Creates Graphs for a Country
###### Includes total, aviation, industry, and power
class CountryGraph:
    def __init__(self, name = None, code = None, df = None):
        self.name = name
        self.code = code
        self.df = df

  #creates graph with given dataframe from excel sheet
    def create_total_graph(self, name = None, code = None):
        Country_Name = str(self.name)
        Country_Code = str(self.code)

        self.df = pd.read_excel('PublicData_Countries_v6 - Updated.xlsx', sheet_name = Country_Code, usecols=("D,F,M,T"))
        self.df.columns = ["Time Since January 1st", "Total CO2, Median", "Total CO2, Low", "Total CO2, High"]

        pd.set_option('display.max_rows', self.df.shape[0] + 1)
        pd.set_option('display.max_columns', self.df.shape[0] + 1)

        plt.title(f"Change in CO2 emissions per day {nl}{Country_Name} since January 1st, 2020", fontsize = 30, fontname = "Times New Roman")
        plt.xlabel("Time Since January 01, 2020 (Days)", fontsize = 35, fontname = "Times New Roman")
        plt.ylabel("Change in CO2 Emissions (Metric Tons)", fontsize = 30, fontname = "Times New Roman")

        #plt.plot('Time Since January 1st', 'Total CO2, Low', data=self.df, marker='o', markerfacecolor='red', markersize=2, color='red', linewidth=2, linestyle = 'dashed')
        plt.plot('Time Since January 1st', 'Total CO2, Median', data=self.df, marker='o', markerfacecolor = 'black', markersize = 2, color='black', linewidth=2)
        #plt.plot('Time Since January 1st', 'Total CO2, High', data=self.df, marker='o', markerfacecolor = 'green', markersize = 2, color='green', linewidth=2, linestyle = 'dashed')

        plt.grid()

  #creates graph of aviation emissions data from excel sheet
    def aviation_graph(self, name = None, code = None):
        Country_Name = str(self.name)
        Country_Code = str(self.code)

        self.df = pd.read_excel('PublicData_Countries_v6 - Updated.xlsx', sheet_name = Country_Code, usecols=("D,L,S,Z"))
        self.df.columns = ["Time Since January 1st", 'CO2 Caused by Aviation, Median', 'CO2 Caused by Aviation, Low', 'CO2 Caused by Aviation, High']
        pd.set_option('display.max_rows', self.df.shape[0] + 1)
        pd.set_option('display.max_columns', self.df.shape[0] + 1)

        plt.title(f"Change in CO2 emissions per day in {Country_Name} since January 1st, 2020 due to Aviation")
        plt.xlabel("Time Since January 01, 2020 (Days)", fontsize = 20, fontname = "Times New Roman")
        plt.ylabel("Change in CO2 Emissions (Metric Tons)", fontsize = 20, fontname = "Times New Roman")

        #plt.plot('Time Since January 1st', 'CO2 Caused by Aviation, Low', data=self.df, marker='o', markerfacecolor='red', markersize=2, color='red', linewidth=2, linestyle = 'dashed')
        plt.plot('Time Since January 1st', 'CO2 Caused by Aviation, Median', data=self.df, marker='o', markerfacecolor = 'blue', markersize = 2, color='blue', linewidth=2)
        #plt.plot('Time Since January 1st', 'CO2 Caused by Aviation, High', data=self.df, marker='o', markerfacecolor = 'green', markersize = 2, color='green', linewidth=2, linestyle = 'dashed')

        plt.grid()

  #creates graph of industrial emissions data from excel sheet
    def industry_graph(self, name = None, code = None):
        Country_Name = str(self.name)
        Country_Code = str(self.code)

        self.df = pd.read_excel('PublicData_Countries_v6 - Updated.xlsx', sheet_name = Country_Code, usecols=("D,H,O,V"))
        self.df.columns = ["Time Since January 1st", 'CO2 Caused by Industry, Median', 'CO2 Caused by Industry, Low', 'CO2 Caused by Industry, High']

        pd.set_option('display.max_rows', self.df.shape[0] + 1)
        pd.set_option('display.max_columns', self.df.shape[0] + 1)

        plt.title(f"Change in CO2 emissions per day in {Country_Name} since January 1st, 2020 due to Industry")
        plt.xlabel("Time Since January 01, 2020 (Days)", fontsize = 20, fontname = "Times New Roman")
        plt.ylabel("Change in CO2 Emissions (Metric Tons)", fontsize = 20, fontname = "Times New Roman")

        #plt.plot('Time Since January 1st', 'CO2 Caused by Industry, Low', data=self.df, marker='o', markerfacecolor='red', markersize=2, color='red', linewidth=2, linestyle = 'dashed')
        plt.plot('Time Since January 1st', 'CO2 Caused by Industry, Median', data=self.df, marker='o', markerfacecolor = 'brown', markersize = 2, color='brown', linewidth=2)
        #plt.plot('Time Since January 1st', 'CO2 Caused by Industry, High', data=self.df, marker='o', markerfacecolor = 'green', markersize = 2, color='green', linewidth=2, linestyle = 'dashed')

        plt.grid()

  #creates graph of power emissions data from excel sheet
    def power_graph(self, name = None, code = None):
        Country_Name = str(self.name)
        Country_Code = str(self.code)

        self.df = pd.read_excel('PublicData_Countries_v6 - Updated.xlsx', sheet_name = Country_Code, usecols=("D,G,N,U"))
        self.df.columns = ["Time Since January 1st",  'CO2 Caused by Power, Median', 'CO2 Caused by Power, Low','CO2 Caused by Power, High']

        pd.set_option('display.max_rows', self.df.shape[0] + 1)
        pd.set_option('display.max_columns', self.df.shape[0] + 1)

        #plt.title(f"Change in CO2 emissions per day in {Country_Name} since January 1st, 2020 due to Power") 
        plt.xlabel("Time Since January 01, 2020 (Days)", fontsize = 20, fontname = "Times New Roman")
        plt.ylabel("Change in CO2 Emissions (Metric Tons)", fontsize = 20, fontname = "Times New Roman")
        
        #plt.plot('Time Since January 1st', 'CO2 Caused by Power, Low', data=self.df, marker='o', markerfacecolor='red', markersize=2, color='red', linewidth=2, linestyle = 'dashed')
        plt.plot('Time Since January 1st', 'CO2 Caused by Power, Median', data=self.df, marker='o', markerfacecolor = 'yellow', markersize = 2, color='yellow', linewidth=2)
        #plt.plot('Time Since January 1st', 'CO2 Caused by Power, High', data=self.df, marker='o', markerfacecolor = 'green', markersize = 2, color='green', linewidth=2, linestyle = 'dashed')

        plt.grid()
    def more_metrics(self, name = None, code = None):
        Country_Name = str(self.name)
        Country_Code = str(self.code)
        self.df = pd.read_excel('PublicData_Countries_v6 - Updated.xlsx', sheet_name = Country_Code, usecols=("D,I,J,K"))
        self.df.columns = ["Time Since January 1st", 'CO2 Caused by Surface Transportation, Median', 'CO2 Caused by the Public Sector, Median', "CO2 Caused by the Residential Sector, Median"]
        plt.xlabel("Time Since January 01, 2020 (Days)", fontsize = 20, fontname = "Times New Roman")
        plt.ylabel("Change in CO2 Emissions (Metric Tons)", fontsize = 20, fontname = "Times New Roman")
        plt.plot('Time Since January 1st', 'CO2 Caused by Surface Transportation, Median', data=self.df, marker='o', markerfacecolor = 'yellow', markersize = 2, color='purple', linewidth=2)
        plt.plot('Time Since January 1st', 'CO2 Caused by the Public Sector, Median', data=self.df, marker='o', markerfacecolor = 'yellow', markersize = 2, color= 'orange', linewidth=2)
        plt.plot('Time Since January 1st', 'CO2 Caused by the Residential Sector, Median', data=self.df, marker='o', markerfacecolor = 'yellow', markersize = 2, color='green', linewidth=2)

        
        
        
    def all_graphs(self, name = None, code = None):
        
        plt.figure(figsize=(20,10))
        plt.xticks(np.arange(0, 160, step=10))
        if self.code == "USA" or self.code == "CHN":
            plt.yticks(np.arange(-7, 2, step=.25))
        elif self.code == "TKM":
            plt.yticks(np.arange(-1,0,step=.001))
        else:
            plt.yticks(np.arange(-7, 2, step=.10))

        plt.xlabel("Time Since January 01, 2020 (Days)", fontname = "Times New Roman", fontsize = 20)
        plt.ylabel("Change in CO2 Emissions (Metric Tons)", fontsize = 20, fontname = "Times New Roman")
        plt.grid()
        self.aviation_graph(name, code)
        self.power_graph(name, code)
        self.more_metrics(name, code)
        self.industry_graph(name, code)
        self.create_total_graph(name, code)
        plt.legend()
        plt.grid(linewidth = 1)
        plt.savefig(f'{self.name} emissions data')
        plt.show()

    def __str__ (self, name = None):
        print(self.name)
        print(self.code)


########## DRIVER CODE ##########
#CA = USAStateGraph("California", 'usa_ca')
#CA.create_graph()
UK = CountryGraph("United Kingdom", "GBR")
CHINA = CountryGraph("China", "CHN")
#CHINA.all_graphs()

USA = CountryGraph("United States", "USA")
SPAIN = CountryGraph("Spain", "ESP")
ITALY = CountryGraph("Italy", "ITA")
SOUTH_KOREA = CountryGraph("South Korea", "KOR")
BRAZIL = CountryGraph("Brazil", 'BRA')
AUSTRALIA = CountryGraph("Australia", "AUS")
MEXICO = CountryGraph("Mexico", "MEX")
TURKMENISTAN = CountryGraph("Turkmenistan", "TKM")
country_list = [UK, CHINA, SPAIN, ITALY, SOUTH_KOREA, BRAZIL, AUSTRALIA, MEXICO, USA, TURKMENISTAN]
for item in country_list:
    item.all_graphs()


```
