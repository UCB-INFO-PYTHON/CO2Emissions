```
import numpy as np
import pandas as pd
import random 
import matplotlib.pyplot as plt
class USAStateGraph:
  def __init__(self, name = None, code = None, df = None):
    self.name = name
    self.code = code
    self.df = df
  def create_graph(self, name = None, code = None):
    State_Name = str(self.name)
    State_Code = str(self.code)
    self.df = pd.read_excel(r'/content/PublicData_USA_States_v6 (1) - Updated.xlsx', sheet_name = State_Code, usecols=("D,F,M,T"))
    self.df.columns = ["Time Since January 1st", "Total CO2, Medium", "Total CO2, Low", "Total CO2, High"]
    pd.set_option('display.max_rows', self.df.shape[0] + 1)
    pd.set_option('display.max_columns', self.df.shape[0] + 1)
    plt.title(f"Change in CO2 emissions per day in {State_Name} since January 1st, 2020")
    plt.xlabel("Time Since January 01, 2020 (Days)")
    plt.ylabel("Change in CO2 Emissions (Metric Tons)")
    plt.plot('Time Since January 1st', 'Total CO2, Low', data=self.df, marker='o', markerfacecolor='red', markersize=2, color='red', linewidth=2, linestyle = 'dashed')
    plt.plot('Time Since January 1st', 'Total CO2, Medium', data=self.df, marker='o', markerfacecolor = 'orange', markersize = 2, color='orange', linewidth=2, linestyle = 'dashed')
    plt.plot('Time Since January 1st', 'Total CO2, High', data=self.df, marker='o', markerfacecolor = 'green', markersize = 2, color='green', linewidth=2, linestyle = 'dashed')
    plt.grid()
    plt.show()
  def __str__ (self, name = None):
    print(self.name)
    print(self.code)

class CountryGraph:
  def __init__(self, name = None, code = None, df = None):
    self.name = name
    self.code = code
    self.df = df
  def create_graph(self, name = None, code = None):
    Country_Name = str(self.name)
    Country_Code = str(self.code)
    self.df = pd.read_excel(r'/content/PublicData_Countries_v6 - Updated.xlsx', sheet_name = Country_Code, usecols=("D,F,M,T"))
    self.df.columns = ["Time Since January 1st", "Total CO2, Medium", "Total CO2, Low", "Total CO2, High"]
    pd.set_option('display.max_rows', self.df.shape[0] + 1)
    pd.set_option('display.max_columns', self.df.shape[0] + 1)
    plt.title(f"Change in CO2 emissions per day in {Country_Name} since January 1st, 2020")
    plt.xlabel("Time Since January 01, 2020 (Days)")
    plt.ylabel("Change in CO2 Emissions (Metric Tons)")
    plt.plot('Time Since January 1st', 'Total CO2, Low', data=self.df, marker='o', markerfacecolor='red', markersize=2, color='red', linewidth=2, linestyle = 'dashed')
    plt.plot('Time Since January 1st', 'Total CO2, Medium', data=self.df, marker='o', markerfacecolor = 'orange', markersize = 2, color='orange', linewidth=2, linestyle = 'dashed')
    plt.plot('Time Since January 1st', 'Total CO2, High', data=self.df, marker='o', markerfacecolor = 'green', markersize = 2, color='green', linewidth=2, linestyle = 'dashed')
    plt.grid()
    plt.show()
  def __str__ (self, name = None):
    print(self.name)
    print(self.code)
UK = CountryGraph("United Kingdom", "GBR")
UK.create_graph()
CHINA = CountryGraph("China", "CHN")
CHINA.create_graph()
SPAIN = CountryGraph("Spain", "ESP")
SPAIN.create_graph()
ITALY = CountryGraph("Italy", "ITA")
ITALY.create_graph()
SOUTH_KOREA = CountryGraph("South Korea", "KOR")
SOUTH_KOREA.create_graph()
BRAZIL = CountryGraph("Brazil", 'BRA')
BRAZIL.create_graph()
AUSTRALIA = CountryGraph("Australia", "AUS")
AUSTRALIA.create_graph()
MEXICO = CountryGraph("Mexico", "MEX")
MEXICO.create_graph()
TURKMENISTAN = CountryGraph("Turkmenistan", "TKM")
TURKMENISTAN.create_graph()



```
